```python

def run_quiz():
    print("\n--- Começando o Quiz de Carros! ---")

    # Sortear 10 perguntas do banco
    selected_questions = random.sample(questions, 10)

    score = 0

    for i, q in enumerate(selected_questions):
        print(f"\nPergunta {i + 1}/10:")
        print(q["question"])
        print(f"a) {q["option_a"]}")
        print(f"b) {q["option_b"]}")

        user_answer = input("Sua resposta (a ou b): ").strip().lower()

        while user_answer not in ['a', 'b']:
            print("Resposta inválida. Por favor, digite 'a' ou 'b'.")
            user_answer = input("Sua resposta (a ou b): ").strip().lower()

        if user_answer == q["correct_answer"]:
            print("Correto! Muito bem!")
            score += 1
        else:
            print(f"Errado! A resposta correta era '{q["correct_answer"]}'.")
            print("\n--- Fim de jogo! Você perdeu. ---")
            return # Termina o jogo imediatamente

    if score == 10:
        print("\n--- Parabéns! Você acertou todas as 10 perguntas e GANHOU o jogo! ---")

# Para iniciar o quiz, execute a função:
run_quiz()
```
